## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
y = np.array([1, 4, 7, 12, 14, 19, 23, 27, 56, 57])

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, p(x))
plt.scatter(x, y)



# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```
