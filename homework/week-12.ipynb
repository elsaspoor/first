## Histogram (3 points)

```python
# Create a histogram from random data

import numpy as np
import seaborn as sns
import pandas as pd

data = np.random.multivariate_normal([0,0], [[1,0], [0,1]], size=10)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
  sns.histplot(data[col], alpha=1)
```
